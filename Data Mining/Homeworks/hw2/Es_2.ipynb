```
from pyspark.sql import SparkSession
import numpy as np
import matplotlib.pyplot as plt

def compute_distance(point, centroid, metric):
    if metric == 'L2':
        return float(np.sum((point - centroid) ** 2))
    elif metric == 'L1':
        return float(np.sum(np.abs(point - centroid)))
    else:
        raise ValueError("Metric must be either 'L1' or 'L2'")

def kmeans_clustering(data_df, centroid_df, metric='L2', k=10, max_iter=20, init="C1"):
    spark = SparkSession.builder.appName("KMeansClustering").getOrCreate()

    data_rdd = data_df.rdd.map(lambda row: np.array([float(row[i]) for i in range(len(row))]))
    centroid_rdd = centroid_df.rdd.map(lambda row: np.array([float(row[i]) for i in range(len(row))]))

    centroids = centroid_rdd.collect()

    errors = []

    for iteration in range(max_iter):
        # Broadcast centroids to all workers
        bc_centroids = spark.sparkContext.broadcast(centroids)

        # Assign each point to the closest centroid
        clustered_data = data_rdd.map(lambda point: (
            np.argmin([compute_distance(point, c, metric) for c in bc_centroids.value]), point
        ))

        # Calculate error for the current iteration
        current_error = clustered_data.map(lambda x: compute_distance(x[1], centroids[x[0]], metric)).sum()
        errors.append(current_error)

        # Calculate new centroids: Group points by assigned cluster, then compute their mean
        # Explanation:
        # - `map(lambda x: (x[0], (x[1], 1)))`: Create a tuple of (cluster_id, (point, count))
        # - `reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1]))`: Sum all points in each cluster and count the number of points
        # - `mapValues(lambda x: x[0] / x[1])`: Calculate the new centroid by dividing sum of points by count
        # - `sortByKey()`: Ensure the clusters are sorted by their index
        # - `map(lambda x: x[1])`: Extract only the new centroid values from (cluster_id, centroid)
        new_centroids = clustered_data.map(lambda x: (x[0], (x[1], 1))) \
            .reduceByKey(lambda x, y: (x[0] + y[0], x[1] + y[1])) \
            .mapValues(lambda x: x[0] / x[1]) \
            .sortByKey() \
            .map(lambda x: x[1]) \
            .collect()

        if np.array_equal(centroids, new_centroids):
            print(f"\nCentroids have not changed. Stopping the algorithm at iteration {iteration + 1}\n")
            break
        centroids = new_centroids

    err = (errors[0] - errors[9]) / errors[0]
    print(f"\nThe error between iteration 0 and iteration 10 is reduced of {np.round(err * 100, 2)}%")
    print(f"The final error is: {np.round(errors[-1], 2)}\n")

    plt.plot(range(len(errors)), errors, marker='o')
    plt.title(f'K-Means Clustering Error ({metric} distance & {init} Centroids)')
    plt.xlabel('Iteration')
    plt.ylabel('Error')
    plt.grid(True)
    plt.show()

    return centroids, errors
```

```
centroids, errors = kmeans_clustering(data, c1, "L2", 10, 20, "C1")
```
