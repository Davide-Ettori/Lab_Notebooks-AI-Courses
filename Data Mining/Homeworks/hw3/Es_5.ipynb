```python
import numpy as np
```

```python
np.random.seed(42)
x = np.random.randint(0, 100, size=(15, 10))

print("Matrix x:\n\n", x)

U, S, VT = np.linalg.svd(x, full_matrices=False)

r = 9

document_embeddings = np.dot(U[:, :r], np.diag(S[:r]))

word_embeddings = np.dot(np.diag(S[:r]), VT[:r, :])

document_embeddings = np.round(document_embeddings, 2)
word_embeddings = np.round(word_embeddings, 2)

print("\n\nDocument Embeddings:\n\n", document_embeddings)
print("\n\nWord Embeddings:\n\n", word_embeddings.T)
print("\n")
print("Shape of Document Embeddings:", document_embeddings.shape)
print("Shape of Word Embeddings:", word_embeddings.T.shape)
```
