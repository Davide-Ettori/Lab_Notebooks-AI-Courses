```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
W1 = np.array([
    [1, -1],
    [-1, -1],
    [0, -1]
])
b1 = np.array([1, 1, -1])
W2 = np.array([1, 1, -1])
b2 = -1.5
step = np.vectorize(lambda x: 1 if x >= 0 else 0)

def run_nn(X):
    res = list()
    for vec in X:
        res.append(step(W2 @ step(W1 @ vec + b1) + b2))
    return np.array(res)
```

```python
X = np.random.uniform(-2, 2, size=(1000, 2))
y = run_nn(X)
```

```python
blue = list()
red = list()

for i, point in enumerate(X):
    if y[i] == 0:
        blue.append(point)
    elif y[i] == 1:
        red.append(point)

blue = np.array(blue)
red = np.array(red)

plt.scatter(blue[:, 0], blue[:, 1], c='blue')
plt.scatter(red[:, 0], red[:, 1], c='red');
```

```python
a, b, c = -0.6, 0, 0.9
curve_x = np.linspace(-2, 2, 500)
curve_y = a * curve_x ** 2 + b * curve_x + c

plt.scatter(blue[:, 0], blue[:, 1], c='blue')
plt.scatter(red[:, 0], red[:, 1], c='red')
plt.scatter(curve_x, curve_y, c='green');
```
